```python
# Beat tracking example
from __future__ import print_function
import librosa
import librosa.display
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import IPython.display as ipd
```

```python
import numpy as np
import pandas as pd
```

```python
SR = 22050
```

### LOAD DATASET

```python
X_mel = np.load('data/features/X_tonnetz.npy')
```

```python
X_mel.shape
```

## chew-bite

```python
S = X_mel[0,:,:]
```

```python
plt.figure(figsize=(10, 4))
#S_dB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S, x_axis='time', y_axis='tonnetz', sr=SR,
                          fmax=SR/2)
plt.colorbar()
plt.ylabel('Tonal Centroids (Tonnetz)')
plt.title('Spectral contrast')
plt.tight_layout()
plt.show()
```
